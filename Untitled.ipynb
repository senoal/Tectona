```python
import pandas as pd
```

```python
a = pd.DataFrame(
    {
        "key1": ["K1", "K12", "K13", "K14", "K15"],
        "key2": ["K21", "K22", "K22", "K23", "K24"],
        "A": ["A0", "A1", "A2", "A3", "A4"],
        "B": ["B0", "B1", "B2", "B3", "A5"],
    }
)
```

```python
a
```

```python
b = pd.DataFrame(
    {
        "key1": ["K1", "K12", "K13", "K14"],
        "key2": ["K21", "K22", "K22", "K22"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
b
```

```python
result = pd.merge(a, b, on=["key1", "key2"])
result
```

```python
kiri = pd.merge(a, b, how="left", on=["key"])
kiri
```

```python
kanan = pd.merge(a, b, how="right", on=["key"])
kanan
```
